```python
data=[]

for file in files:
    
    if file[-4:] == '.csv':
        
        data.append(file)
```

```python
data[0].split()[0]
```

```python
standard_deviations = []
scale_factor = .001

for file in data:
    
    table = pd.read_csv(file)
    name = file.split()[0]
    
    table['Scaled X Values'] = scale_factor*table['x values']
    
    y_data = table['y values'].values
    deviations = np.std(y_data)
    standard_deviations.append(deviations)
    
    plt.scatter(table['Scaled X Values'],table['y values'],label=name)

x_axis = np.arange(-1*scale_factor,101*scale_factor,.1*scale_factor)
plt.plot(x_axis,2*x_axis/scale_factor,label='Theoretical')

plt.xlabel('x')
plt.ylabel('y ($\Omega$)')

plt.title('All data')

plt.legend()
```
