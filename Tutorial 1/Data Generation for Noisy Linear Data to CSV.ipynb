```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
number_of_datapoints = 100
```

```python
data = stats.uniform().rvs(number_of_datapoints)*100 
noise = np.random.normal(0,15,100)
```

```python
dependent_variable = 2*data + noise
```

```python
columns_for_table = {'x values':data,'y values': dependent_variable}
```

```python
generated_data = pd.DataFrame(columns_for_table)
generated_data
```
